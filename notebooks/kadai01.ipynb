課題1) 下記座標群から、線形回帰のモデルを作成せよ

```python
import numpy as np
import matplotlib.pyplot as plt

b_train = -1
w_train = 0.7
x_train = np.random.random((1, 100))
y_train = x_train * w_train + b_train + 0.1 * np.random.randn(1, 100)

plt.figure(1)
plt.plot(x_train, y_train, 'ro', label='Data')
plt.show()
```
